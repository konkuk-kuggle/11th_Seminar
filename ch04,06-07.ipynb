# 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
tips=sns.load_dataset("tips") #식당에서 팁을 지불한 손님의 정보 모아둔 데이터.
print(tips.head())
print(type(tips))
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
```

```python
#히스토그램 그래프 만들기(변수 1개)

axes1.hist(tips['total_bill'],bins=10) #total bill을 변수로 가지는 히스토그램. bins의 인자값은 x축의 간격을 나타냄.
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

```python
#산점도 그래프 만들기(변수 2개)

scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip']) #scatter 메서드에 total_bill, tips 열 전달하면 산점도 그래프 완성.
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
#박스 그래프

boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1) #틀만기기

axes1.boxplot([tips[tips['sex']=='Female']['tip'], #첫번째 인자는 tips 데이터프레임에서 성별이 Female인 데이터와 Male인 데이터에서 tip 열 데이터만 추출하여 리스트로 담아 전달.
               tips[tips['sex']=='Male']['tip']], 
              labels=['Female','Male']) #두번째 인자는 labels 인잣값으로 성별을 구분하기 위함.

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
#다변량 그래프 그리기
#다변량 그래프: '3개 이상'의 변수를 사용한 그래프.

print(tips)

def recode_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1

tips['sex_color']=tips['sex'].apply(recode_sex) #sex 열에 recode_sex함수를 브로드캐스팅하기 위해 apply 메서드 사용. 10장에서 설명.
```

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10, #배수 안적으면 점들이 너무 작음. 숫자가 커질수록 사이즈 커짐.
    c=tips['sex_color'], 
    alpha=0.5) #alpha 인자값으로 점의 투명도 조절. 숫자 커질수록 불투명해짐.
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

# 단변량 그래프 그리기 - 히스토그램

```python
import seaborn as sns #seaborn 라이브러리는 matplotlib라이브러리보다 좀 더 화려한 그래프 그릴 수 있음.

tips=sns.load_dataset("tips")
```

```python
#seaborn 라이브러리로 히스토그램 그리려면:subplots, distplot메서드 활용.

ax=plt.subplots() #기본틀 만들기.
ax=sns.distplot(tips['total_bill']) #열데이터 전달하면 히스토그램 그릴 수 있음./distplot 메서드는 히스토그램과 밀집도 그래프를 같이 보여줌.
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False) #distplot 메서드에서 밀집도 그래프 제외하고 싶다면 kde인자값을 False로 설정하기.
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],hist=False) #밀집도 그래프만 나타내려면 hist인자를 False로 지정하면 됨.
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True) #rug인자를 추가함으로써 양탄자(rug)그래프 까지 그림
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
#양탄자 그래프란 그래프의 축에 동일한 길이의 직선을 붙여 '데이터의 밀집 정도'를 표현한 그래프임.
```

```python
#히스토그램과 비슷하게 생긴 count 그래프 그려보기!

ax=plt.subplots()
ax=sns.countplot(x='day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
#seaborn 라이브러리로 산점도 그래프 그리기-regplot메서드 사용

ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips) #seaborn 라이브러리의 regplot 메서드 사용해 산점도 그래프 그리는 법(이변량그래프).
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips, fit_reg=False) #산점도 그래프에서 회귀선을 제거하려면 fit_reg 인자를 False로 지정하면 됨.
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#산점도 그래프와 히스토그램을 한 번에 그려주는 jointplot 메서드.

joint=sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#산점도 그래프는 점이 겹칠 경우 구분하기 어려움. 육각그래프(hexbin)를 사용하면 구분하기 쉬워짐.
#특정데이터 개수가 많아지면 점점 진한색으로 표현. jointplot 메서드를 그대로 사용하고, 대신 kind 인잣값을 hex로 지정하면 됨.

hexbin=sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#이차원 밀집도 그리기!-kdeplot 메서드

ax=plt.subplots()
ax=sns.kdeplot(x=tips['total_bill'],
               y=tips['tip'],
               shade=True) #음영 효과 지정
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#바 그래프 그리기!- barplot 메서드 사용/바 그래프는 지정한 변수의 '평균'을 계산하여 그릴 수 있음.

ax=plt.subplots()
ax=sns.barplot(x='time',y='total_bill',data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#박스 그래프 그리기!-boxplot 메서드 사용

ax=plt.subplots()
ax=sns.boxplot(x='time',y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
#바이올린 그래프- 박스 그래프에서 데이터의 분산을 쉽게 확인할 수 있는 그래프.

ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
#관계 그래프 그리기! - pairplot 메서드/ 지금까지의 모든 그래프를 종합한 그래프라 생각하면 됨.
#메서드 안에 데이터프레임을 넣으면 간단하게 그려짐.

fig=sns.pairplot(tips)
```

```python
#우하향 대각선 기준으로 위는 나머지 절반과 대칭. 중복된 그래프는 위치 직접 지정후 원하는 그래프로 교체 가능-map_upper or map_lower

pair_grid=sns.PairGrid(tips) #틀 지정
pair_grid=pair_grid.map_upper(sns.regplot) #대각선 위로는 산점도 그래프
pair_grid=pair_grid.map_lower(sns.kdeplot) #대각선 아래로는 이차원 밀집도
pair_grid=pair_grid.map_diag(sns.distplot, rug=True) #대각선은 히스토그램, 밀집도
plt.show()
```

# 누락값 확인하기

```python
#누락값은 NaN,NAN,nan과 같은 방법으로 표기 가능. NaN이라고 표기하겠음.
#누락값을 파이썬에 사용하려면 numpy(넘파이) 라이브러리가 필요함.(수학이나 과학 연산을 위해 만든 파이썬 라이브러리임.)

from numpy import NaN,NAN,nan #누락값 불러오기

print(NaN==True) #누락값은 0,''와 같은 개념이 아니라, '데이터 자체가 없다'는 뜻임.
print(NaN==False)
print(NaN==0)
print(NaN=='')

print(NaN==NaN)
print(NaN==nan)
print(NaN==NAN)
print(nan==NAN)
```

# 누락값을 포함한 데이터를 불러올 때

```python
import pandas as pd #판다스에는 누락값을 확인하는 메서드인 isnull이 있음.

print(pd.isnull(NaN))
print(pd.isnull(nan))
print(pd.isnull(NAN))

print(pd.notnull(NaN)) #반대의 경우(즉 누락값이 아닌 경우)는 notnull 메서드를 사용하면 됨.
print(pd.notnull(42))
print(pd.notnull('missing'))
```

# 인덱스를 다시 만들 때 누락값이 발생하는 경우

# 누락값의 개수 구하기

```python
ebola=pd.read_csv('../data/country_timeseries.csv')

print(ebola.count()) #count메서드는 '누락값이 아닌 값의 개수'를 구하는 메서드.

print(ebola.shape[0]) #shape[0]에는 전체 행 데이터 개수가 저장되어 있음.

num_rows=ebola.shape[0]
num_missing=ebola.shape[0]-ebola.count()
print(num_missing) #변수 별로 누락값의 개수를 출력함.
```

```python
import numpy as np

print(np.count_nonzero(ebola.isnull())) #count_nonzero는 배열에서 0이 아닌 값의 개수를 세는 메서드임.

print(np.count_nonzero(ebola['Cases_Guinea'].isnull())) #count_nonzero와 isnull 조합해도 누락값 개수 구할 수 있음.
```

```python
print(ebola.Cases_Guinea.value_counts(dropna=False).head()) 
#value_counts는 지정한 열의 빈도를 구하는 메서드임. '누락값의 개수'를 구하려면 dropna=False 전달하면 됨.
```

# 누락값을 다른 값으로 변경하기

```python
#누락값 처리하기 - 변경

print(ebola.fillna(0).iloc[0:10, 0:5]) #fillna 메서드에 0을 대입하면 누락값을 0으로 변경함!
```

```python
print(ebola.fillna(method='ffill').iloc[0:10, 0:5]) #fillna 메서드의 method 인자값에 ffill을 지정하게 되면 누락값이 나타나기 전의 행에 있는 값으로 누락값 변경.
#6행의 누락값은 5행의 값을 사용해 그대로 8018.0의 값을 가지게 됨. 그러나 0,1행은 처음부터 누락값이므로 누락값 그대로 사용.
```

```python
print(ebola.fillna(method='bfill').iloc[0:10, 0:5])
#method인자에 bfill 지정할 경우 누락값이 나타나고 난 바로 다음 행에 있는 값으로 누락값 변경. 즉 ffill의 반대방향. 마지막값이 누락값일 경우 처리못함.
```

```python
print(ebola.interpolate().iloc[0:10, 0:5]) #누락값 양쪽에 있는 값을 이용하여 중간값을 구한 다음 누락값 처리.
```

# ebola 데이터 집합 살펴보기

```python
ebola=pd.read_csv('../data/country_timeseries.csv')
print(ebola.columns)
print(ebola.iloc[:5,[0,1,2,3,10,11]])
```

```python
ebola_long=pd.melt(ebola, id_vars=['Date','Day'])
print(ebola_long.head()) #각나라별 사망자 수를 행으로 볼 수 있음.
```

# 열 이름 나누고 데이터 프레임에 추가하기

```python
#문자를 기준으로 열 분리하기-split 메서드 사용하기(기본적으로 공백을 기준으로 문자열 자름.)

variable_split=ebola_long.variable.str.split('_')
print(variable_split)

print(type(variable_split))
print(type(variable_split[0]))
```

```python
status_values=variable_split.str.get(0) #get 메서드를 사용해 0번째 인덱스 추출
country_values=variable_split.str.get(1) #get 메서드를 사용해 1번째 인덱스 추출

print(status_values[:5])
print(status_values[-5:])
print(country_values[:5])
print(country_values[-5:])
```

```python
ebola_long['status']=status_values
ebola_long['country']=country_values
print(ebola_long.head())
```

# concat 메서드를 응용하여 데이터프레임에 열 추가하기

```python
#concat 메서드를 활용하면 split 메서드로 분리한 데이터를 바로 데이터프레임에 추가할 수 있음.

variable_split=ebola_long.variable.str.split('_',expand=True)
variable_split.columns=['status','country']
ebola_parsed=pd.concat([ebola_long,variable_split], axis=1)

print(ebola_parsed) #status, country 열이 한번씩 더 추가됨.
```

# 기상 데이터의 여러 열을 하나로 정리하기

```python
weather=pd.read_csv('../data/weather.csv')
print(weather.iloc[:5,:]) 
```

```python
weather_melt=pd.melt(weather, id_vars=['id','year','month','element'],var_name='day',value_name='temp')
print(weather_melt.head()) #피벗해도 최고,최저 온도가 한눈에 잘 들어오지는 않음.
```

```python
weather_tidy=weather_melt.pivot_table( #피벗테이블 메서드는 행과 열의 위치를 다시 바꿔 정리해줌.
    index=['id','year','month','day'], #index인자에는 위치를 그대로 유지할 열 이름을 지정
    columns='element', #columns 인자에는 피벗할 열 이름을 지정.
    values='temp', #values 인자에는 새로운 열의 데이터가 될 열 이름을 지정.
    dropna=False
)
print(weather_tidy)
```

```python
weather_tidy_flat=weather_tidy.reset_index() #새로운 데이터프레임의 인덱스를 reset_index 메서드로 지정한 것.

print(weather_tidy_flat.head())
```

# 빌보드 차트의 중복 데이터 처리하기

```python
billboard=pd.read_csv('../data/billboard.csv')
billboard_long=pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'], var_name='week', value_name='rating')

print(billboard_long.shape)

print(billboard_long.head())
```

```python
print(billboard_long[billboard_long.track=='Loser'].head()) #노래제목이 Loser인것만 따로 모았음.
```

```python
billboard_songs=billboard_long[['year','artist','track','time']] #중복데이터를 가지고 있는 열들.
print(billboard_songs.shape)
```

```python
billboard_songs=billboard_songs.drop_duplicates() #drop_duplicates메서드는 데이터프레임의 중복 데이터를 제거함.
print(billboard_songs.shape)
```

```python
billboard_songs['id']=range(len(billboard_songs)) #id를 추가하면 데이터를 쉽게 조회하거나 구분할 수 있음.
print(billboard_songs.head(n=10))
```

```python
billboard_ratings=billboard_long.merge(billboard_songs, on=['year','artist','track','time']) #merge 메서드를 사용해 노래정보와 주간순위데이터를 합침.
print(billboard_ratings.shape)
print(billboard_ratings.head())
```
